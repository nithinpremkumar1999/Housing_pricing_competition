```python
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')

train_data.dropna(subset=['SalePrice'], inplace=True, axis=0)
y = train_data.SalePrice
train_data.drop(['SalePrice'], axis=1, inplace=True)

numeric_cols = [col for col in train_data.columns if train_data[col].dtype in ['float64', 'int64']]

X_train = train_data[numeric_cols].copy()
X_test = test_data[numeric_cols].copy()
```

```python
X_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import cross_val_score

def get_score(n_estimators):
    my_pipeline = Pipeline(steps=[
        ('preprocessor', SimpleImputer()),
        ('model', RandomForestRegressor(n_estimators, random_state=0))
    ])
    scores = -1 * cross_val_score(my_pipeline, X_train, y,
                                cv=5, scoring='neg_mean_absolute_error')
    return scores.mean()
```

```python
results = {}
for i in range(1, 9):
    results[50*i] = get_score(50*i)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(list(results.keys()), list(results.values()))
plt.show()
```

```python
my_pipeline = Pipeline(steps=[
    ('preprocessor', SimpleImputer()),
    ('model', RandomForestRegressor(n_estimators=200, random_state=0))
])
my_pipeline.fit(X_train, y)
```

```python
test_preds = my_pipeline.predict(X_test)
output = pd.DataFrame({
    'Id': X_test.index,
    'SalePrice': test_preds
})
output.to_csv('Cross-Validation - Submission')
```
